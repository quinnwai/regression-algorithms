```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "./data/Concrete_Data.xls"

raw_df = pd.read_excel(path)
raw_df.head()
```

```python
for i in raw_df.columns:
    print(i.split(" ("))
```

```python
raw_df.columns = [col.split("(")[0].strip() for col in raw_df.columns]
raw_df.describe()
```

```python
for col in raw_df.columns:
    plt.figure()
    plt.hist(raw_df[col])
    plt.title(f"{col} Histogram")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.savefig(f"./figures/histogram_{col.replace(' ', '_')}")
    plt.show()
```

```python
for col in raw_df.columns[:-1]:
    plt.hist(raw_df[col], alpha=0.8, label=col)
    plt.title(f"Histogram of All Features")

plt.xlabel("Values")
plt.ylabel("Count")
plt.legend()
plt.savefig(f"./figures/histogram_by_features")
plt.show()
```
